# Language Model Module

First prepare the language models of all chatrooms

```python
import LM

all_LMs = LM.preparation('../../Data/FB/JN_msgDict.txt')  # select a facebook chat data to perform LM matching
```

Get the top-10 similar chatrooms of given chatting pair to see the performance.

(currently LM.py only support inner-product distance)

```python
ret = LM.similar_chat(all_LMs, '捷恩蔡', '羅強', show_top=10)
```

LM.similar_chat will return the sorted list of chatrooms according to the similarity to query chatroom in descending order

to query with group chat:

```python
ret = LM.similar_chat(all_LMs, "doesn't matter", "what I put here", show_top=10, alter='張培軒, 捷恩蔡, Ting-Fang Chien')
```
